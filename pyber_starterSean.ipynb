```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
city_data_complete = pd.merge(ride_data, city_data, how="left", on=["city", "city"])

# Display the data table for preview
city_data_complete.to_csv("data/fileOne.csv", index=False, header=True)
```

## Total Fares by City Type

```python
# Calculate Type Percents
urban_fare = 39854.38 / city_data_complete["fare"].sum() *100
suburban_fare = 19356.33 / city_data_complete["fare"].sum() *100
rural_fare = 4327.93 / city_data_complete["fare"].sum() *100
# Build Pie Chart
city_type = ["Urban", "Suburban", "Rural"]
fare_percent = [urban_fare, suburban_fare, rural_fare]
colors = ["gold", "lightblue", "lightcoral"]
explode = (0,0,0)
plt.pie(fare_percent, explode=explode, labels=city_type, colors=colors, autopct="%1.1f%%", shadow=True, startangle=140)

# Save Figure
plt.savefig("data/PyPieFare.png")
```

```python
total_ride = city_data_complete["type"].count()
total_ride
```

## Total Rides by City Type

```python
# Calculate Ride Percents
rural_ride = 125 / total_ride *100
suburban_ride = 625 / total_ride * 100
urban_ride = 1625 / total_ride *100
# Build Pie Chart
ride_type = ["Urban", "Suburban", "Rural"]
ride_percent = [urban_ride, suburban_ride, rural_ride]
colors = ["gold", "lightblue", "lightcoral"]
explode = (0,0,0)
plt.pie(ride_percent, explode=explode, labels=ride_type, colors=colors, autopct="%1.1f%%", shadow=True, startangle=140)

# Save Figure
plt.savefig("data/PyPieRide.png")
```
